# **1. Descriptive Analysis**

```python
#import relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Diabetes Model/diabetesalcohol.csv")
```

```python
#overall dataset info
print(df.info())
print(df.describe(include='all'))
```

```python
#Diabetes rate by alcohol group
print(df.groupby('alcohol_consumption_category')['type2_diabetes'].agg(['count', 'sum', 'mean']))
```

```python
#Crosstab: Diabetes by gender and alcohol group
print(pd.crosstab(
    index = [df['gender'], df['alcohol_consumption_category']],
    columns = df['type2_diabetes'],
    margins=True
    ))
```

```python
#Crosstab: Diabetes by gender and ethnicity
print(pd.crosstab(
    index = [df['gender'], df['ethnicity']],
    columns = df['type2_diabetes'],
    margins=True
    ))
```

```python
#Diabetes by Education Status
print(pd.crosstab(df['education'], df['type2_diabetes'], margins=True))
```

```python
#Distribution plot for alcohol consumption (g/week)
plt.figure()
sns.histplot(df['weekly_alcohol_grams'], bins=30, kde=True)
plt.title('Distribution of Alcohol Consumption (g/week)')
plt.xlabel('Weekly Alcohol Consumption (g)')
plt.ylabel('Count')
plt.show()
```

```python
#Boxplot of BMI by diabetes outcome
plt.figure()
sns.boxplot(x='type2_diabetes', y='bmi', data=df)
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('Type 2 Diabetes')
plt.ylabel('BMI')
plt.show()
```

```python
#Bar Chart: Diabetes Prevalence by Alcohol Consumption Category
sns.barplot(x='alcohol_consumption_category', y='type2_diabetes', data=df)
plt.ylabel('Diabetes Prevalence')
plt.xlabel('Alcohol Consumption Category')
plt.title('Diabetes by Alcohol Consumption Category')
plt.show()
```
